```python
#Top products
```

```python
group = df.groupby(['StockCode','Description'])
top_product = group['Revenue'].sum().sort_values(ascending=False)
top_product.head(10)
```

```python
# Calculating frequency
frequency = grouping['InvoiceNo'].nunique()
frequency
```

```python
# Calculating monetory
monetory = grouping['Revenue'].sum()
monetory
```

```python
# change the date string to datetime type
reffdate = pd.Timestamp('2011-12-10')
```

```python
# Calculating Recency
day = grouping['InvoiceDate'].apply(lambda x :(reffdate - x.max()).days)
day
```

```python
# Create a rfm table using Id , frequency , monetary , recency
rfm = grouping.agg({'InvoiceNo':'nunique' , 'Revenue':'sum' ,'InvoiceDate':(lambda x :(reffdate - x.max()).days)})
rfm.columns = ['frequency','monetary','recency']
rfm
```

```python
# Create recency score 
```

```python
rfm['rscore'] = pd.qcut(rfm['recency'],5,labels =['5','4','3','2','1'])
rfm
```

```python
# Create monetary score 
```

```python
rfm['mscore'] = pd.qcut(rfm['monetary'],5,labels =['1','2','3','4','5'])
rfm
```

```python
# Create frequency score 
```

```python
rfm['fscore'] = pd.qcut(rfm['frequency'].rank(method='first'),5,labels =['1','2','3','4','5'])
rfm
```

```python
# Creating score column combining three score column
```

```python
rfm['score']= (rfm['rscore'].astype(str) + rfm['fscore'].astype(str) + rfm['mscore'].astype(str) )
rfm
```

```python
# Data analysis on the basis of rfm score
```

```python
lowfreq = rfm[rfm['fscore'].astype(int) <2]
lowfreq
```

```python
mostrecent = rfm[rfm['rscore'].astype(int) >4]
mostrecent
```

```python
highmonetary = rfm[rfm['mscore'].astype(int) > 4]
highmonetary
```

```python
champions = rfm[rfm['score'] =='555']
champions
```

```python
atrisk = rfm[rfm['score'] =='311']
atrisk
```

```python
# To satisfy multiple conditions on multiple column
result = rfm[(rfm['fscore'].astype(int) >4) & (rfm['mscore'].astype(int) == 4 )]
result
```
